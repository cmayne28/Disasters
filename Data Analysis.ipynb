```python
import pandas as pd 
import matplotlib.pyplot as plt
import gmaps
```

```python
#Bar Graph - Deaths by Decade (Eruptions)
df = pd.read_csv('Data/v_deaths_decade.csv')

plt.bar(df['Decade'], df['Total Deaths'], color='red')
plt.xticks(size=8)
plt.ylabel("Total Deaths")
plt.xlabel("Decade")
plt.title('Total Volcanic Eruptions Deaths by Decade')
plt.savefig('Images/Eruption_Deaths_by_Decade.png')
plt.show()
```

```python
#Bar Graph - Deaths by Decade (EQs)

df1 = pd.read_csv('Data/e_deaths_decade.csv')

plt.bar(df1['Decade'], df1['Total Deaths'], color='blue')
plt.xlabel('Decades', size=10)
plt.ylabel('Number of Deaths', size=10)
plt.xticks(size=8)
plt.title('Total Earthquake Deaths by Decade')
plt.savefig('Images/Eq_Deaths_by_Decade.png')
plt.show()
```

```python
#Bar Graph - Top 5 Countries with Most Volcanic Eruptions 

df3 = pd.read_csv('Data/total_erupts_country.csv')

#Create Bar Chart
plt.bar(df3["Country"].head(), df3["Total"].head())
plt.xlabel("Country")
plt.ylabel('Total Eruptions')
plt.title("Countries Most Affected by Volcanic Eruptions")
plt.savefig('Images/Top_Countries_Eruptions.png')
plt.show()
```

```python
#Pie Chart - Top 5 Countries with Most Volcanic Eruptions 

explode=[0.1,0,0,0,0]
plt.pie(df3["Total"].head(),labels=df3["Country"].head(),explode=explode, autopct='%1.1f%%')
plt.axis("equal")
plt.show()
#need to change the colors of this
```

```python
#Bar Graph - Top 5 Countries with Most EQs
#Plot Eruptions per decade

df4 = pd.read_csv('Data/total_eqs_country.csv')

plt.bar(df4["COUNTRY"].head(), df4["Total"].head(), color='lawngreen')
plt.xlabel("Country")
plt.ylabel('Total Earthquakes')
plt.title("Countries Most Affected by Earthquakes")
plt.savefig('Images/Top_Countries_EQs.png')
plt.show()
```

```python
#Pie Chart - Top 5 Countries with Most EQs

explode=[0.1,0,0,0,0]
colors = ['cornflowerblue', 'deepskyblue','b', 'mediumslateblue', 'cyan']
plt.pie(df4["Total"].head(),labels=df4["COUNTRY"].head(),explode=explode, colors=colors, autopct='%1.1f%%')
plt.axis("equal")
plt.savefig('Images/Top_Countries_Eqs_Pie.png')
plt.show()
```

```python
#Eruptions & Earthquakes by Decade

df5 = pd.read_csv('Data/total_erupts_decade.csv')
df6 = df5.reindex([8,7,11,10,9,5,6,2,3,0,4,1]).reset_index(drop=True)

df7 = pd.read_csv('Data/total_eqs_decade.csv')
df8 = df7.reindex([8,10,9,6,11,7,4,5,3,2,0,1]).reset_index(drop=True)

plt.plot(df8["DECADE"], df8["Total"], label= "Earthquakes")
plt.plot(df6["Decade"], df6["Total"], label="Volcanic Eruptions")
plt.title("Total Significant Disasters by Decade")
plt.xlabel("Decade")
plt.ylabel("Total Number of Events")
plt.xticks(size=8)
plt.legend()
plt.savefig('Images/Total_Disasters_by_Decade.png')
plt.show()
```

```python
#Earthquakes - Damage by Magnitude Scatter Plot

eq_df = pd.read_csv('earthquakes.csv')

plt.scatter(eq_df["EQ_PRIMARY"], eq_df["TOTAL_DAMAGE_MILLIONS_DOLLARS"])
plt.title('Earthquake Magnitude vs Damage')
plt.xlabel('Magnitude')
plt.ylabel('Damage ($Millions)')
plt.savefig('Images/EQ_Mag_Damage.png')
plt.show()
```
