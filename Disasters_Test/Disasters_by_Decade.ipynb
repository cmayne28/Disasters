```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
file = 'eruptions_by_decade.csv'
df = pd.read_csv(file)

del df["Unnamed: 0"]
df

df["Decade"] = df["index"]

del df["index"]

df.head()
```

```python
file = 'eqs_by_decade.csv'
df1 = pd.read_csv(file)

df1

df2 = df1[["DECADE", "Total Earthquakes"]]

df2.head()
```

```python
new_df = df.merge(df2, left_on='Decade', right_on='DECADE', how='outer')

new_df

new_df1 = new_df[["Decade", "Total Earthquakes", "Total Eruptions"]]

new_df1
```

```python
plt.plot(new_df1["Decade"], new_df1["Total Earthquakes"])
plt.plot(new_df1["Decade"], new_df1["Total Eruptions"])
plt.title("Total Significant Disasters by Decade")
plt.xlabel("Decade")
plt.ylabel("Total Number of Events")
plt.figure(figsize=(18, 18))
plt.show()
```
